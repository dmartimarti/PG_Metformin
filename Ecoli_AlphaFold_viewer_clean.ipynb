# AlphaFold E. coli protein viewer

This notebook is a small application to easily view and explore any protein from _E. coli_ predicted by AlphaFold 2 ([Jumper _et al_., 2021, Nature](https://www.nature.com/articles/s41586-021-03819-2)). 
The proteins are downloaded from the [EMBL-EBI AlphaFold](https://alphafold.ebi.ac.uk/) repository, upzipped and then represented with [py3DMol](http://3dmol.csb.pitt.edu/).

The download speed is limited by the EMBL bandwith, which can vary greatly between runs. So, please, be patient if it takes a bit. 

To run the colab, just click on the "play" buttons on the left of each field, and wait for it to finish. You can also click on Runtime -> Run all to run a example of the code. 

If you wish to represent another protein, don't run all the code again, just select another protein from the list, and run again that cell. 

# Data preparation

```python
#@title

print('Downloading latest database from EMBL-EBI/AlphaFold 2\n')
# data preparation
!wget -q https://ftp.ebi.ac.uk/pub/databases/alphafold/UP000000625_83333_ECOLI.tar --no-check-certificate
print('Installing some necessary dependencies\n')
# install dependencies
!pip -q install py3Dmol
!pip -q install biopandas

# load libraries
import pandas as pd
import numpy as np
import sys
import gzip
import shutil

from biopandas.pdb.engines import pdb_records
from biopandas.pdb.engines import pdb_df_columns
from biopandas.pdb.engines import amino3to1dict
from biopandas.pdb import PandasPdb

import py3Dmol
import tarfile
import glob

# extract tar into folder
my_tar = tarfile.open('UP000000625_83333_ECOLI.tar')
my_tar.extractall('./Ecoli_v1') # specify which folder to extract to
my_tar.close()

!mkdir ./Ecoli_v1/pdb

print('Extracting all PDB files into a subfolder\n')
for f in glob.glob("/content/Ecoli_v1/*pdb.gz"):
  out_name = f.split('/')[-1][1:-3]
  out_name ="/content/Ecoli_v1/pdb/" + out_name
  with gzip.open(f, 'rb') as f_in:
      with open(out_name, 'wb') as f_out:
          shutil.copyfileobj(f_in, f_out)
print('Extraction finished!\n')

paths = glob.glob("/content/Ecoli_v1/pdb/*pdb")

# this creates useful dictionaries to plot the proteins
pdb_files = []
prots = []
for path in paths:
  pdb_file = path.split('/')[-1]
  pdb_files.append(pdb_file)

for file in pdb_files:
  prot = file.split('-')[1]
  prots.append(prot)

# create the dictionary from both datasets
prot_dict = dict(zip(prots, pdb_files))


# function to represent the protein

def prot_viewer(protein):
  print(f'Showing protein {protein}')
  with open("/content/Ecoli_v1/pdb/"+prot_dict[protein]) as ifile:
    system = "".join([x for x in ifile])

  chA = {'chain':'A'}

  view = py3Dmol.view(width=900, height=800)
  view.addModelsAsFrames(system)
  view.setStyle({'model': -1}, {"cartoon": {'color': 'spectrum'}})
  # view.setStyle({'cartoon': {'colorscheme': {'prop':'b_factor','gradient': 'roygb','min':0,'max':1}}})
  # view.addSurface(py3Dmol.VDW,{'opacity':0.4,'color':'white'}, chA)
  view.zoomTo()
  view.show()
```
